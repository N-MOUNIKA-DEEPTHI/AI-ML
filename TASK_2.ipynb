# Perform Exploratory Data Analysis on the provided dataset to extract meaningful insights using Pandas, Matplotlib/Plotly

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('weatherHistory.csv')
data
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Display basic information about the dataset
print(data.info())
```

```python
# Display summary statistics
print(data.describe(include='all'))
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values in each column:")
print(missing_values, "\n")

```

```python
data.columns
```

#EDA(EXPLORATORY DATA ANALYSIS)

```python
# Check for unique values in categorical columns specific column
column = 'Precip Type'
unique_values = data[column].unique()
print(unique_values)
```

```python
# Check for unique values in categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    unique_values = data[column].unique()
    print(f"{column}: {unique_values}\n")
```

```python
#calculate corelation  matrix
# Convert columns to numeric
numeric_columns = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)',
       'Wind Bearing (degrees)', 'Visibility (km)', 'Loud Cover',
       'Pressure (millibars)', 'Daily Summary']
for column in numeric_columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')
# Check for correlations between numerical variables
correlations = data[numeric_columns].corr() # Calculate correlations on the DataFrame subset
print("\nCorrelation matrix of numerical variables:")
print(correlations)
```

#VISUALIZATION

```python
# Visualize missing values
plt.figure(figsize=(12, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Set up the matplotlib figure for data distribution
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

# Temperature (C)
sns.histplot(data['Temperature (C)'], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Temperature (C) Distribution')

# Apparent Temperature (C)
sns.histplot(data['Apparent Temperature (C)'], bins=30, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Apparent Temperature (C) Distribution')

# Humidity
sns.histplot(data['Humidity'], bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Humidity Distribution')

# Wind Speed (km/h)
sns.histplot(data['Wind Speed (km/h)'], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Wind Speed (km/h) Distribution')

# Visibility (km)
sns.histplot(data['Visibility (km)'], bins=30, kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Visibility (km) Distribution')

# Pressure (millibars)
sns.histplot(data['Pressure (millibars)'], bins=30, kde=True, ax=axes[2, 1])
axes[2, 1].set_title('Pressure (millibars) Distribution')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Visualize relationships between variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature (C)', y='Apparent Temperature (C)', data=data)
plt.title('Temperature vs Apparent Temperature')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Humidity', y='Temperature (C)', data=data)
plt.title('Humidity vs Temperature')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Wind Speed (km/h)', y='Wind Bearing (degrees)', data=data)
plt.title('Wind Speed vs Wind Bearing')
plt.show()
```

```python
# Pairplot to visualize relationships
sns.pairplot(data[['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Visibility (km)', 'Pressure (millibars)']])
plt.show()
```
